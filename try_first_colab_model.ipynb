```python
import spacy
import os
import pandas as pd
import csv
from spacy import displacy

import datetime
now = datetime.datetime.now()
print ("Current date and time : ", now.strftime("%Y-%m-%d %H:%M:%S"))

# load mode
nlp = spacy.load('/content/drive/MyDrive/complete_model/Complete_model')
print('Model loaded from : ', '/content/drive/MyDrive/complete_model/Complete_model')
text= []
while text not in ['done', 'DONE', 'Done', 'close', 'Close', 'CLOSE', 'EXIT', 'Exit', 'exit']:
    text= input('Enter you text here : ')
    doc = nlp(text)
    print(text)
    for ent in doc.ents:
        print(ent.label_, ent.text)
    displacy.render(nlp(doc.text), style='ent', jupyter=True)
```
